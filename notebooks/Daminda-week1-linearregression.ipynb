```python
y_mean = df[['TARGET_5Yrs']].mean()
y_mean
```

```python
y_shape = (len(df), 1)
y_base = np.full(y_shape, y_mean)
y_base
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
print(mse(df['TARGET_5Yrs'], y_base, squared=False))
print(mae(df['TARGET_5Yrs'], y_base))
```

```python
y = df[['TARGET_5Yrs']]
```

```python
X = df.drop(['TARGET_5Yrs', 'Id'], axis=1)
```
